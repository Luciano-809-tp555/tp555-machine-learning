```python
# Import all necessary libraries
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import matplotlib
from pylab import *
import numpy as np
import random

%matplotlib notebook
#%matplotlib widget
```

```python
# Number of pairs feature/label.
N = 1000
```

```python
# Input values (features)
x = np.random.randn(N, 1)

# Output values (targets).
y = 1.0 + np.power(x, 2) + np.random.randn(N, 1)
```

```python
# Plot features (input) versus targets (output)
fig = plt.figure(figsize=(10,10))
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,'.')
plt.show() 
```

```python
# Generate values for parameters.
a0 = np.linspace(-30.0, 30.0, 100)
a1 = np.linspace(-30.0, 30.0, 100)

# Generate points for plotting the cost-function surface.
J = np.zeros((100,100))
for iter0 in range(0, 100):
    
    for iter1 in range(0, 100):
        
        yhat = a0[iter0] + a1[iter1]*x
    
        J[iter0][iter1] = (1/N)*np.sum( np.square(y - yhat)  );

# Plot cost-function surface.
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')

A0, A1 = np.meshgrid(a0, a1)

surf = ax.plot_surface(A0, A1, J, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_xlabel('a0')
ax.set_ylabel('a1')
ax.set_zlabel('J');

plt.title('Cost-function\'s Surface')

ax.view_init(20, 45)
fig

#Show the plot.
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))

cp = plt.contour(A0, A1, J)
plt.clabel(cp, inline=1, fontsize=10)
plt.xlabel('a0')
plt.ylabel('a1')
plt.title('Cost-function\'s Contour')
plt.xticks(np.arange(-30, 32, step=2.0))
plt.yticks(np.arange(-30, 32, step=2.0))
plt.show()
```
