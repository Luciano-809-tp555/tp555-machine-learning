```python
# Import all necessary libraries
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import matplotlib
from pylab import *
import numpy as np
import random
```

```python
N = 100

# Input values (features)
x = 10.*np.random.randn(1, N)

# Output values (targets).
y = 0.2 + 3.*x + np.random.randn(1, N)

# Generate values for parameters.
a0 = np.linspace(-10.0, 10.0, 100)
a1 = np.linspace(-10.0, 10.0, 100)

J = np.zeros((100,100))
for iter0 in range(0, 100):
    
    for iter1 in range(0, 100):
        
        yhat = a0[iter0] + a1[iter1]*x
    
        J[iter0][iter1] = (1/N)*np.sum( np.square(y - yhat)  );
        

    
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')

A0, A1 = np.meshgrid(a0, a1)

surf = ax.plot_surface(A0, A1, J, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_xlabel('a0')
ax.set_ylabel('a1')
ax.set_zlabel('J');

ax.view_init(45, 90)
fig

#Show the plot
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
#ax = fig.gca(projection='3d')

cp = plt.contour(A0, A1, J)
plt.clabel(cp, inline=1, fontsize=10)
plt.xlabel('A0')
plt.ylabel('A1')
plt.show()
```
