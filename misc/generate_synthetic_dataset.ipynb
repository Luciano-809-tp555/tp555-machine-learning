```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import matplotlib

# Generate a synthetic dataset from a quadratic function
N = 1000 #Number of data points
# Generate random x values between -5 and 5

vec = []
for i in range(0,N):
    vec.append( np.random.rand() )
vec = sorted(vec)

v = np.zeros((N, 1))
for i in range(0,N):
    v[i] = vec[i]

x = 10*( v - 0.5  )

# Define the function and the true parameters
# y = a_0 + a_1*x + a_2*x^2
a_0 = 1;
a_1 = -2;
a_2 = 0.5;

# Define t
y = a_0 + a_1*x + a_2*(np.power(x, 2) )

# Add some noise
y = y + 0.5*np.random.randn(N, 1)

i = 0
writer = csv.writer(open("sythquad.csv", 'w'), delimiter=',')
for row in y:
    writer.writerow((x[i][0], y[i][0]))
    i = i + 1
    

fig = plt.figure(figsize=(10,10))
plt.plot(x, y, 'b.')
    
print('Ready!')
```
