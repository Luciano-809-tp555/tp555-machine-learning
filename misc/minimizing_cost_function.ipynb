```python
# Import all necessary libraries
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import matplotlib
from pylab import *
import numpy as np
import random

#%matplotlib notebook
#%matplotlib widget
```

```python
N = 100

# Input values (features)
x = 10.0*np.random.rand(N, 1)

# Output values (targets).
y = 0.2 + 3.0*x + np.random.randn(N, 1)

# Plot features (input) versus targets (output)
fig = plt.figure(figsize=(10,10))
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,'.')
plt.show() 
```

```python
# Generate values for parameters.
a0 = np.linspace(-10.0, 10.0, 100)
a1 = np.linspace(-10.0, 10.0, 100)

A0, A1 = np.meshgrid(a0, a1)

J = np.zeros((100,100))
for iter0 in range(0, 100):
    
    for iter1 in range(0, 100):
        
        yhat = a0[iter0] + a1[iter1]*x
    
        J[iter0][iter1] = (1/N)*np.sum( np.power((y - yhat), 2)  );
        

# Plot Error surface.        
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')

surf = ax.plot_surface(A0, A1, J, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_xlabel('a0')
ax.set_ylabel('a1')
ax.set_zlabel('J');

ax.view_init(0, 90)
fig

#Show the plot
plt.show()
```
